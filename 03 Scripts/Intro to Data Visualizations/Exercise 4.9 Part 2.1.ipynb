## Part 2, step 3 -  Create a histogram of the “order_hour_of_day” column

```python
# 3 Create a histogram of the “order_hour_of_day” column.
df_merged['order_hour_of_day'].plot.hist(bins=24)
```

```python
# Answer to 3.1: The histogram shows that most orders occur during daytime, especially between 9am and 4pm, peaking around 10–11am, 
# while orders drops sharply during late night and early morning hours. This suggests that late morning to early afternoon is the most popular ordering time.
```

```python
# 4 Creating a bar chart from loyalty_flag column
df_merged['loyalty_flag'].value_counts().plot.bar()
```

```python
# 5 Checking whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day with a line chart
# Notes: To keep the line chart clear, I grouped data by hour and used average prices instead of raw values.
df_merged.groupby('order_hour_of_day').agg({'prices': 'mean'}).plot.line()
```

```python
# 6 Creating a line chart comparing age and number of dependents. 
df_merged.groupby('age').agg({'num_dependents': 'mean'}).plot.line()
```

```python
# Answer to 6.1: 
# The line chart shows the average number of dependents for customers at each age. 
# The values fluctuate slightly without a clear trend, suggesting that the number of dependents doesn’t strongly vary by age in this dataset. 
# This visualization helps identify customer lifecycle stages, which can guide more targeted marketing.
```
